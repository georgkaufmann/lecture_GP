In this notebook, we will discuss and then read data files (topography, Total-field anomaly) 
saved in the `geodyn5` format.

<img src="images/FUGeocampus_topo_boug.png" style="height:15cm">

First, we load some `python` libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy
import cartopy.crs as ccrs
import libGP
#libGP.change_fontsize()
```

----
## Reading geodyn5 data files

In this first part, we read the field data (elevation, Bouguer anomaly, ...), using the functions in the `libGP`
library.

```python
easting,northing,elevation = libGP.readTopography("2Dm_GP2024_DTM.xy",path='./data/FUGeocampus/',iskip=5,control=True,plot=True)
```

```python
totaleast,totalnorth,totaltopo,totaldata = libGP.readTotalField('2Dm_GP2021_MAG_Rigole.xy',path='data/FUGeocampus/',iskip=1,control=True,plot=True)
```

----
## Read DOP image
We then read the **digital orthophoto** (DOP), using the `imread` function from `matplotlib`.

Note that the image is read as pixel image, see the output of the `plt.imshow` command.

We therefore need to **re-project** the pixel dimensions to real coordinates, which is done wit the
`img_extent` field, containing min/max values of the easting and northing coordinates.

```python
imgname = "data/FUGeocampus/dop20rgbi_33_388_5808_2_be_2023.jp2"
img_extent = (388000,390000, 5808000,5810000)
img = plt.imread(imgname)
plt.imshow(img)
```

----
## Read structure data
Next, we read in the digitised outlines from the **tunnel** and the **rigole**, which we digitised
within the GIS software.

As both digitised outlines are polygons, we can use `plt.fill` to plot filled polygons.

```python
Tunnel = np.loadtxt("data/FUGeocampus/Tunnel.xy",skiprows=1)
print(Tunnel.shape)
Rigole = np.loadtxt("data/FUGeocampus/Rigole.xy",skiprows=1)
print(Rigole.shape)

plt.fill(Tunnel[:,0],Tunnel[:,1],color='red',alpha=0.3)
plt.fill(Rigole[:,0],Rigole[:,1],color='blue',alpha=0.3)
```

----
## Plotting Plotting total-field anomaly on topography and DOP
Next, we plot the elevation and field data ...

Note that we use the `cartopy` package from `python`,  which reprojects the coordinates to the UTM reference system,
here the zone 33, in which Berlin is located.

```python
plt.figure(figsize=(10,10))
ax = plt.axes(projection=ccrs.UTM(zone="33"),aspect='equal')
plt.title('FU Geocampus, Berlin\n Orthophoto with topography')
# draw ticks and labels
ax.set_xlim([388300,388500])
ax.set_xticks([388300,388400,388500])
ax.set_xlabel('Easting [m]')

ax.set_ylim([5809700,5809900])
ax.set_yticks([5809700,5809800,5809900])
ax.set_ylabel('Northing [m]')
ax.ticklabel_format(useOffset=False)

# DOP image
ax.imshow(img, origin='upper', extent=img_extent,transform=ccrs.UTM(zone="33"))

# Tunnel and rigole
plt.fill(Tunnel[:,0],Tunnel[:,1],label='Tunnel',color='red',alpha=0.3,transform=ccrs.UTM(zone="33"))
plt.fill(Rigole[:,0],Rigole[:,1],label='Rigole',color='blue',alpha=0.3,transform=ccrs.UTM(zone="33"))
plt.plot(Tunnel[:,0],Tunnel[:,1],color='red',lw=1,ls='dashed',transform=ccrs.UTM(zone="33"))
plt.plot(Rigole[:,0],Rigole[:,1],color='blue',lw=1,ls='dashed',transform=ccrs.UTM(zone="33"))

# faint topo image
vmin=41;vmax=46
bounds = np.linspace(vmin,vmax,11)
cbar1=ax.tricontourf(easting,northing,elevation,cmap='terrain',vmin=vmin,vmax=vmax,levels=bounds,alpha=0.2,
                     transform=ccrs.UTM(zone="33"))
plt.colorbar(cbar1,label='Elevation [m]',shrink=1.0,location='bottom')

# topo contours
cbar2=ax.tricontour(easting,northing,elevation,levels=bounds,cmap='terrain',linewidths=1,alpha=0.6,
                    transform=ccrs.UTM(zone="33"))
#ax.clabel(cbar2,colors='black',inline=True,fmt='{:.0f}'.format)

# total data
cbar2=ax.tricontourf(totaleast,totalnorth,totaldata,cmap='seismic',alpha=0.80,
                     transform=ccrs.UTM(zone="33"))
plt.colorbar(cbar2,label='Total-field anomaly [nT]',shrink=0.9,location='right')

#currentXvalues = plt.gca().get_xticks()
#currentYvalues = plt.gca().get_yticks()
plt.gca().set_xticklabels(['{:.0f}'.format(x) for x in plt.gca().get_xticks()])
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in plt.gca().get_yticks()])
plt.legend()
plt.savefig("images/FUGeocampus_topo_total.png")
```
