```python
import numpy as np
import matplotlib.pyplot as plt

import libGP
```

----
## Reading data

### Read **elevation** data in `geodyn` format

```python
help(libGP.readTopography)
```

```python
easting,northing,elevation = libGP.readTopography('2Dm_GP2022_DTM.xy',path='data/',iskip=20,control=True,plot=True)
```

----
### Read **Bouguer anomaly** in `geodyn5` format

```python
help(libGP.readBouguerAnomaly)
```

```python
bougeast,bougnorth,bougtopo,bougdata = libGP.readBouguerAnomaly('2Dm_GP2018_GRAV_BK11.xy',path='data/',iskip=1,control=True,plot=True)
```

----
### Read **Total-field anomaly** in `geodyn5` format

```python
help(libGP.readTotalField)
```

```python
totaleast,totalnorth,totaltopo,totaldata = libGP.readTotalField('2Dm_GP2022_MAG_M01.xy',path='data/',iskip=1,control=True,plot=True)
```

----
### Read **ERT profile** in `geodyn5` format

```python
help(libGP.readERTprofile)
```

```python
ert,points,tri = libGP.readERTprofile('2Dc_GP2022_ERT02_Wenner.xz',path='data/',control=True,plot=True)
```

----
### Read **GPR profile** in `geodyn5` format

```python
help(libGP.readGPRprofile)
```

----
## Create **coordinate** files

```python
help(libGP.createERTCoordElevation)
```

```python
nElectrodes = 50
sElectrodes = 5
GPSPoints = np.array([
[572637.000,5667728.000],
[572413.000,5667667.000]
])

elecPoints = libGP.createERTCoordElevation('GP2022_ERT01_WenSchlum',nElectrodes,sElectrodes,GPSPoints,easting,northing,elevation,
                                           path='tmp/',control=True,plot=True)
```

```python
help(libGP.createGPRCoordElevation)
```

```python
lProfile = 78
sProfile = 1
traceInc = 0.01
GPSPoints = np.array([
[572637.000,5667728.000],
[572413.000,5667667.000]
])

gprPoints = libGP.createGPRCoordElevation('GP2022_GPR01',lProfile,sProfile,GPSPoints,easting,northing,elevation,traceInc,
                                           path='tmp/',control=True,plot=True)
```

```python
help(libGP.createGravityCoordElevation)
```

```python
libGP.createGravityCoordElevation('BlaueKuppe.xy',path='tmp/',irepeat=3,control=True,plot=True)
```

----
## Add **profile length and elevation** to **ERT data** in `Res2DInv` format

```python
help(libGP.addERTCoordElevation)
```

```python
libGP.addERTCoordElevation('GP2022_ERT01_WenSchlum.dat',elecPoints,path='tmp/',control=False)
```
