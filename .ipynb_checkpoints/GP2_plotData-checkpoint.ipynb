```python
import numpy as np
import matplotlib.pyplot as plt
import libGP
```

----
## Reading data
We use the following function from `libGP` to plot data sets:
- `readTopography`
- `readBouguerAnomaly`
- `readTotalField`
- `readERTprofile`
- `readGPRprofile`

### Read **elevation** data in `geodyn` format

```python
help(libGP.readTopography)
```

```python
easting,northing,elevation = libGP.readTopography("2Dm_GP2024_DTM.xy",path='data/FUGeocampus/',iskip=5,control=True,plot=True)
```

----
### Read **Bouguer anomaly** in `geodyn5` format

```python
help(libGP.readBouguerAnomaly)
```

```python
bougeast,bougnorth,bougtopo,bougdata = libGP.readBouguerAnomaly('2Dm_GP2021_GRAV_Rigole.xy',path='data/FUGeocampus/',iskip=1,control=True,plot=True)
```

----
### Read **Total-field anomaly** in `geodyn5` format

```python
help(libGP.readTotalField)
```

```python
totaleast,totalnorth,totaltopo,totaldata = libGP.readTotalField('2Dm_GP2021_MAG_Rigole.xy',path='data/FUGeocampus/',iskip=1,control=True,plot=True)
```

----
### Read **ERT profile** in `geodyn5` format

```python
help(libGP.readERTprofile)
```

```python
ert,points,tri = libGP.readERTprofile('2Dc_GP2021_ERT02_Schlumberger.xz',path='data/FUGeocampus/',control=True,plot=True)
```

----
### Read **GPR profile** in `geodyn5` format

```python
help(libGP.readGPRprofile)
```

```python
gpr,points,tri = libGP.readGPRprofile('2Dc_GP2021_GPR02.xz',path='data/FUGeocampus/',iskip=10,control=True,plot=True,scale=1e5)
```
