```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,20,2)
print(x)
x=np.append(x,[x+1])
print(x)
```

```python
y = np.sin(x/10*2*np.pi)
print(x)
print(y)
```

```python
plt.plot(x,y,lw=0,marker='o',markersize=10,color='red')
plt.plot(x,y,lw=2,color='blue')
```

```python
print(np.sort(x))
```

```python
print(np.argsort(x))
```

```python
sorted = np.argsort(x)
print(x[sorted],y[sorted])
```

```python
sorted = np.argsort(x)
plt.plot(x[sorted],y[sorted],lw=0,marker='o',markersize=10,color='red')
plt.plot(x[sorted],y[sorted],lw=2,color='blue')
```
